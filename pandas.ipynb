```python
# List of 5 Anime Titles
anime = ['Attack on Titan', 'My Hero Academia', 'One Piece', 'Death Note', 'Fullmetal Alchemist: Brotherhood']

# List of 5 Anime Genres
genre = ['Shounen', 'Superhero', 'Adventure', 'Psychological', 'Fantasy']
anime_series=pd.Series(anime, name='anime genre')
```

```python
anime_series
```

```python
genre_series=pd.Series(genre,index=anime_series, name='anime genre')
genre_series
```

```python
# List of 5 Anime Titles, we can also pass tjis as a numpy array
import numpy as np
anime =np.array(['Attack on Titan', 'My Hero Academia', 'One Piece', 'Death Note', 'Fullmetal Alchemist: Brotherhood'])

# List of 5 Anime Genres we can pass this as a tuple
genre = ('Shounen', 'Superhero', 'Adventure', 'Psychological', 'Fantasy')
anime_series=pd.Series(anime,index=genre, name='anime genre')
anime_series
```

```python
#dictionary
dec={
'Shounen'          :                 'Attack on Titan',
'Superhero'        :                      'One Piece',
'Psychological'     :                     'Death Note',
'Fantasy'        :  'Fullmetal Alchemist: Brotherhood'
}
```

```python
#creating a dic in series
pd.Series(dec, name='anime genre')
```

```python
#nlargest() and nsmallest 
# Dictionary of Guitars and Prices
guitar_prices = {
    'Fender Stratocaster': 1200.00,
    'Gibson Les Paul': 2200.00,
    'PRS Custom 24': 1800.00,
    'Ibanez RG550': 1000.00,
    'Taylor 814ce': 3000.00,
    'Martin D-28': 2500.00
}

gp=pd.Series(guitar_prices,name='price')
gp
```

```python
gp.nsmallest()
#to kniow the index gp.nsmallest().index[0]
```

```python
gp.nlargest()
```

```python
#index of smallest or largest value
gp.idxmin()
```

```python
gp.idxmax()
```
